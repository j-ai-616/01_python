# 다형성
- 같은 이름의 메서드를 호출하지만 객체의 타입에 따라 서로 다른 결과가 나오는 능력

## 메소드 오버라이딩 (Method Overiding)
- 부모 메소드를 자식이 자기 입맛에 맞게 다시 정의하는 다형성의 핵심 기술

```python
class Animal:
    def speak(self):
        return "..."
    
class Dog(Animal):
    def speak(self):
        return "멍멍"
    
class Cat(Animal):
    def speak(self):
        return "야옹"
    
animals = [Dog(), Cat(), Dog(), Cat(), Animal()]
for animal in animals:
    print(f"{type(animal).__name__} : {animal.speak()}")
```

## 덕 타이(Duck Typing)
"오리처럼 걷고 꽥꽥거리면 오리다." => 상속 관계가 아니더라도 메소드 이름만 같으면 동일하게 취급하는 파이썬의 유연한 특징

```python
class Tiger:
    def speak(self):
        return '어흥'

class Car:
    def speak(self):
        return '빵빵'
    
def make_it_speak(obj):
    # obj가 무엇이든 상관 없이 speak() 메소드만 있으면 실행 가능
    print(obj.speak())
    
make_it_speak(Tiger())
make_it_speak(Car())
```
