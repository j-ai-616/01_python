# Python 환경과 패키지 관리

Python으로 프로그램을 개발할 때는 **프로젝트마다 필요한 라이브러리(패키지)가 서로 다를 수 있다.**

예를 들어 어떤 프로젝트는 `pandas`의 특정 버전을 필요로 하고, 다른 프로젝트는 다른 버전을 요구할 수 있다.  
이러한 상황에서 하나의 Python 환경에 모든 패키지를 설치하면 **버전 충돌(package conflict)** 이 발생할 수 있다.

이 문제를 해결하기 위해 **가상 환경(virtual environment)** 을 사용한다.

가상 환경은 하나의 컴퓨터 안에서 **독립적인 Python 실행 환경을 여러 개 만드는 방식**이다.

```text
내 컴퓨터
├ Python 환경 A (데이터 분석 프로젝트)
├ Python 환경 B (웹 서비스 프로젝트)
└ Python 환경 C (AI 실습 환경)
```

각 환경은 서로 독립적으로 동작하므로, 한 환경에서 설치한 패키지가 다른 환경에 영향을 주지 않는다.

---

# conda 환경

현재 **conda 기반 가상 환경**을 사용하고 있다.

conda는 Python에서 다음 기능을 제공하는 도구이다.

- Python 가상 환경 관리
- 패키지 설치 및 관리
- 라이브러리 의존성 관리

즉, conda는 단순히 패키지를 설치하는 도구라기보다 **Python 환경 자체를 관리하는 도구**이다.

conda에서 패키지를 설치할 때는 **패키지를 가져오는 저장소(repository)** 가 필요한데, 이를 **채널(channel)** 이라고 한다.

대표적인 채널은 다음과 같다.

- `defaults` : Anaconda 기본 채널
- `conda-forge` : 커뮤니티 기반 패키지 저장소

라이센스 이슈를 피하기 위해 **conda-forge 채널을 사용하는 환경**을 사용한다.

conda-forge는 다음 특징이 있다.

- 다양한 패키지 제공
- 업데이트가 빠름
- 데이터 분석 및 과학 계산 라이브러리 호환성이 좋음

# pip와 패키지 설치

Python 패키지는 보통 다음 두 가지 방식으로 설치할 수 있다.

1. `conda install`
2. `pip install`

두 명령어 모두 **Python 패키지를 설치하는 명령어**이지만 역할이 조금 다르다.

| 도구 | 역할 |
|-----|-----|
| conda | Python 환경 + 패키지 관리 |
| pip | Python 패키지 설치 |

---

# PyPI (Python Package Index)

`pip` 명령어로 설치되는 대부분의 Python 패키지는  
**PyPI(Python Package Index)** 라는 저장소에서 제공된다.

PyPI는 Python 패키지를 공개적으로 공유하는 **공식 패키지 저장소(repository)** 이다.

대표적인 예시 패키지

- numpy
- pandas
- requests
- fastapi
- streamlit

`pip install` 명령어를 실행하면 pip는 **PyPI 서버에서 해당 패키지를 다운로드하여 설치한다.**

예시

```bash
pip install requests
```

위 명령어는 다음 과정을 수행한다.

```text
pip 명령 실행
↓
PyPI 서버 접속
↓
requests 패키지 다운로드
↓
현재 Python 환경에 설치
```

# conda 환경에서 pip 사용

우리는 **conda 환경을 사용하지만 일부 패키지는 pip로 설치한다.**

이유는 다음과 같다.

- 모든 Python 패키지가 conda에 등록되어 있는 것은 아니다.
- 일부 라이브러리는 pip(PyPI)에서 먼저 배포되는 경우가 많다.
- 웹 프레임워크나 최신 라이브러리는 pip 설치가 일반적인 경우가 많다.

중요한 점은 **pip 명령어를 사용하더라도 현재 활성화된 conda 환경에 패키지가 설치된다는 점**이다.

예를 들어 다음과 같은 순서로 실행하면

```bash
conda activate pystudy_env
pip install streamlit
```

`streamlit` 패키지는 **pystudy_env 가상 환경 내부에만 설치된다.**

```text
pystudy_env
├ numpy
├ pandas
└ streamlit
```

즉, 다른 Python 환경에는 영향을 주지 않는다.

# 패키지 설치 전략

일반적으로 다음과 같은 방식으로 패키지를 설치한다.

```text
conda 환경 생성
↓
conda install (데이터 분석 라이브러리)
↓
pip install (웹 라이브러리 / 최신 라이브러리)
```

예시

```bash
conda install numpy pandas
pip install streamlit
```
