# 딕셔너리 (Dictionary)
- **키(key)**와 **값(value)**으로 이루어진 자료구조
- 순서보다는 이름(key)을 통해 데이터를 빠르게 찾는 것이 목적

## 딕셔너리의 특징
- 키는 중복 될 수 없으며, 값이 변하지 않는 타입으로 사용

````python
# 생성
teacher = {'name': 'tiger', 'team': 'ohgiraffers', 'age': 40}

# 키의 제약: 리스트는 키가 될 수 없지만 튜플은 가능
# invalid_dict = {['a', 'b'] : 10}  # 리스트(변할 수 있는 타입)는 키로 사용 불가
valid_dict = {('a', 'b'): 10}
print(valid_dict)

# 중복 키: 나중에 선언 된 값이 이전 값을 덮어씀
dup_dict = {'A': 1, 'A': 2}
print(dup_dict)
````

## 데이터의 조회 및 변경

````python
ohgiraffers = {'tiger': 'DB', 'Bear': 'Python', 'Gorilla': 'Spring'}

# 방법 1. 대괄호 사용 (직관적이지만 위험할 수 있음)
print(ohgiraffers['tiger'])
# print(ohgiraffers['fox'])  # KeyError 발생

# 방법 2. get() 메소드 사용 (안전한 프로그래밍)
print(ohgiraffers.get('tiger'))
print(ohgiraffers.get('fox'))  # None 반환 (에러 없음)
print(ohgiraffers.get('fox', '미배정'))  # 찾는 값이 없을 경우, 출력할 내용 지정 가능

# 데이터 추가 및 수정
ohgiraffers['Fox'] = 'React'    # 새로운 키-값 추가
ohgiraffers['tiger'] = 'Java'   # 기존 키의 값 수정

ohgiraffers
````

## 주요 메소드 

````python
# 키, 값, 쌍 추출
print(ohgiraffers.keys())       # 키 목록
print(ohgiraffers.values())     # 값 목록
print(ohgiraffers.items())      # (키, 값) 튜플 쌍 목록
````

## 딕셔너리의 복사 (주의!)
파이썬의 .copy() 는 **얕은 복사(Shallow Copy)** 이다.

### 불변 vs 가변
- 불변 객체(숫자, 문자열, 튜플): 값을 수정하면 아예 새로운 값을 만들어 갈아 끼우기 때문에 복사본에 영향을 주지 않음
- 가변 객체(리스트, 딕셔너리): 내부의 값을 수정해도 '주소'는 그대로 이기 때문에 복사본도 그 변화를 그대로 보게 됨

````python
import copy

# 얕은 복사
a = {'user': {'name': '우별림'}, 'score': 100}
b = a.copy()

a['user']['name'] = '김별림'      # 내부 딕셔너리 수정 (주소 공유 중)
a['score'] = 90                 # 바깥 쪽 숫자 수정 (새로운 숫자로 교체 됨)

print(f"얕은 복사 b: {b}")
print(f"원본 변경 a: {a}")

# 깊은 복사 (deepcopy) -> 내부의 모든 객체까지 완전히 새로 복제
c = {'user': {'name': '우별림'}, 'score': 100}
d = copy.deepcopy(c)

c['user']['name'] = '김별림'  
c['score'] = 90  

print(f"깊은 복사 c: {d}")        # 원본을 수정해도 복사본은 독립적임
print(f"원본 변경 c: {c}")
````
