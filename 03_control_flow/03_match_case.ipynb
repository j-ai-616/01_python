# match...case
- 값 기반 분기문으로 if-elif-else 구조를 더 간결하고 읽기 쉽게 표현할 수 있음
- case _ 는 앞의 어떤 케이스에도 해당하지 않을 때 실행되는 기본 값

```python
# 기본 매칭 (음료 자판기) : 변수의 값이 특정 값과 일치하는지 확인
num = input("음료 선택 : 1. 콜라, 2. 사이다, 3. 쥬스")

match num:
    case "1":
        print("콜라는 1000원입니다.")
    case "2":
        print("사이다는 1200원입니다.")
    case "3":
        print("쥬스는 1500원입니다.")
    case _:
        print("잘못 입력하셨습니다.")

print("이용해주셔서 감사합니다.")
```

```python
# 연산자 분기 처리 (계산기) : 기호에 따라 서로 다른 산술 연산 수행
operand = '!'
a, b = 10, 3

match operand:
    case '+':
        print(f'{a} + {b} = {a + b}')
    case '-':
        print(f'{a} - {b} = {a - b}')
    case '*':
        print(f'{a} * {b} = {a * b}')
    case '/':
        print(f'{a} / {b} = {a / b}')
    case _:
        print(f'부적절한 연산자: {operand}')
```

```python
# 조건문 가드 (if 활용) : case 뒤에 if를 붙여 값의 범위나 추가 조건을 검사
n = int(input('정수 입력:'))

match n:
    case n if n > 0:
        print('양수를 입력하셨습니다.')
    case n if n < 0:
        print('음수를 입력하셨습니다.')
    case _:
        print('0을 입력하셨습니다.')
```

```python
# 구조적 매칭 (튜플 및 딕셔너리) : 데이터의 자료구조 형태완 ㅐ부 값을 동시에 검사
# 튜플 매칭 예시
value = (1, -10)

match value:
    case (1, 2):
        print(f'(1, 2)이 전달 되었습니다.')
    case (1, k) if k < 0:
        print(f'(1, 음수 {k})가 전달 되었습니다.')
    case (1, _):
        print(f'(1, n)가 전달 되었습니다.')
    case _:
        print(f'그 외의 경우')
        
# 딕셔너리 매칭 예시
applicant = {'type': 'dev', 'career': 8}

match applicant:
    case {'type': 'dev', 'career': career} if career >= 10:
        print('중견 개발자 지원입니다.')
    case {'type': 'dev'}:
        print('신입 개발자 지원입니다.')
    case _:
        print('기타 직군 지원입니다.')
```
